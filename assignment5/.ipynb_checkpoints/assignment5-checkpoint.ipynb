# 0. Helper Functions
We begin this assignment by defining the activation functions we will use later on. These include ReLU, Sigmoid, and their respective derivatives when applicable.

```python
#   ReLU activation function
def ReLU (array):
    return np.maximum(0, array)
    
#   derivative of ReLU activation function
def dReLU(array):
    array[array < 0] = 0
    array[array > 0] = 1
    return array

#   Sigmoid activation function
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))
```

We continue by defining a checkAccuracy function that will calculate our accuracy and error rates of our learning models.

```python
#   compute the training and the test error
def checkAccuracy(w, b , X, Y, T, neuralnetworkType) :
    # initialize our paramters for use in forward propogation 
    z = [None] * len(w)
    a = [None] * len(w)

    # forward propopgation 
    # we use a[-1] (the a from the very last layer) to get our prediction labels
    z[0] = w[0] @ X + b[0]
    for i in range(len(w)-1):
        a[i] = ReLU(z[i])
        z[i+1] = w[i+1] @ a[i] + b[i+1]
    a[-1] = sigmoid(z[-1])

    # get the prediction labels  
    result = np.where(a[-1] >= 0.5, 1.0, 0.0)

    # compare our prediction labels to the actual labels (Y in this case)
    dim = result.ndim
    count = 0
    if (dim == 1) :
        for i in range(1, Y.size):
            if result[i] == Y[0, i]:
                count += 1
    else :       
        for i in range(1, Y.size):
            if result[0, i] == Y[0, i]:
                count += 1

    # calculate accuracy and error rates 
    accuracy = (count/Y.shape[1]) * 100
    error_rate = 100 - accuracy

    # print results 
    print(T + " accuracy rate for " + neuralnetworkType + " is : " + str(accuracy) + "%")
    print(T + " error rate for " + neuralnetworkType + " is : " + str(error_rate) + "%")
    print(" ")
```

Finally we define our predict_plot functions to help us plot our decision boundaries later on.

```python
def predictPlotLogisticRegression (w, b , X) : 
    z = w @ X + b
    aTest = sigmoid(z)

    labels = np.array([1 if a > .5 else 0 for a in aTest.T])
    return labels

def predictPlotNeuralNetwork (w, b , X) : 
    z = [None] * len(w)
    a = [None] * len(w)

    # forward propopgation
    z[0] = w[0] @ X + b[0]
    for i in range(len(w)-1):
        a[i] = ReLU(z[i])
        z[i+1] = w[i+1] @ a[i] + b[i+1]
    a[-1] = sigmoid(z[-1])
    labels = np.where(a[-1] >= 0.5, 1.0, 0.0)
    return labels[0,:]
```

### Logistic Regression
Now we use a Logistic Regression model to classify the train set. I designed and implemented this model for assignment 2. We plot the cost function, and also calculate the training and test accuracy and error rates. 

```python
#   Logistic Gradient Descent 
def LGD(X, y, alpha):
    # initialize parameters 
    costs = []
    m = X.shape[1]
    w = np.zeros((X.shape[0], 1))
    b = np.zeros((1, 1))

    for i in range(0,2000):

        # forward propogation
        z = (w.transpose() @ X) + b
        a = 1 / (1 + np.exp(-z))
        dZ = a - y.transpose()

        # compute cost function and append to costs list for later plotting 
        cost = (-1/m) * np.sum(y.transpose() * np.log(a) + (1 - y.transpose()) * np.log(1 - a))
        costs.append(cost)

        # back propogation and update parameters 
        dW = (1/m) * (X @ dZ.transpose())
        dB = (1/m) * np.sum(dZ)
        w = w - (alpha*dW)
        b = b - (alpha*dB)

    # plot cost function
    plt.clf()
    plt.plot(costs)
    plt.title("Logistic Regression Cost Function (" + str(m) + " samples)")
    plt.xlabel("Number of Iterations")
    plt.ylabel("Cost")
    plt.show()

    # return our weights and biases 
    return(w, b)

# train our neural network and get back our weights and biases 
# learning rate alpha = 0.5
w, b = LGD(train_X, train_Y.T, 0.5)

# check the accuracy of our weights and biases 
neuralType = "Logistic Regression"
C = np.ravel(train_Y)
checkAccuracy(w.T, b.T, train_X, train_Y, "Train", neuralType)
checkAccuracy(w.T, b.T, test_X, test_Y, "Test", neuralType)
```

We will get better performance as the complexity of our learning algorithm increases.

# 2. Unregularized Neural Network
We now train a 3-layer neural network to classify the data - it has 2 hidden layers with 20 and 10 neurons respectively, uses ReLU in the hidden layers and sigmoid in the output layer. We plot the cost function, and calculate the training and test accuracy and error rates. 

```python
#   neural network with no regularization
def unregularizedNeuralNetwork(X, y, alpha, layer_list):
    f, m = X.shape
    
    # initialize weights, biases and parameters 
    w = functools.reduce(lambda acc, neurons: (acc[0] + [np.random.randn(neurons, 
    acc[1]) * np.sqrt(2.0/acc[1])], neurons), layer_list, ([], f))[0]
    b = [ np.random.randn(neurons, 1) * 0.01 for neurons in layer_list ]
    z = [None] * len(w)
    a = [None] * len(w)
    costs = []

    for i in range(0, 2000):

        # forward propogation
        z[0] = w[0] @ X + b[0]
        for i in range(len(w)-1):
            a[i] = ReLU(z[i])
            z[i+1] = w[i+1] @ a[i] + b[i+1]
        a[-1] = sigmoid(z[-1])

        # compute cost function and append to costs list for later plotting 
        cost = - (y @ np.log(a[-1].T) + (1 - y) @ np.log(1 - a[-1].T))[0,0] / m
        costs.append(cost)

        dZ = a[-1] - y
        gradients = [] 
        # backward propogation 
        for i in range(len(z)-2, -1, -1):
            # Calculate dW and dB
            gradients.append([ (dZ @ a[i].T) / m, 
                np.sum(dZ, axis=1).reshape(b[i+1].shape) / m ])
        
            dZ = (w[i+1].T @ dZ) * dReLU(z[i])

        # calculate dW and dB for the first layer
        gradients.append([ (dZ @ X.T) / m, 
            np.sum(dZ, axis=1).reshape(b[0].shape) / m ])

        # flip gradients so order is from first to last layer
        gradients = gradients[::-1]

        # update parameters 
        for j in range(len(gradients)):
            w[j] = w[j] - (alpha * gradients[j][0])
            b[j] = b[j] - (alpha * gradients[j][1])

    # plot cost function
    plt.clf()
    plt.plot(costs)
    plt.title("Unregularized Neural Network Cost Function (" + str(m) + " samples)")
    plt.xlabel("Number of Iterations")
    plt.ylabel("Cost")
    plt.show()
    return(w, b)

#    train our neural network and get back our weights and biases 
# learning rate alpha = 0.5, 2 hidden layers with 20 and 10 neurons respectively
w,b = unregularizedNeuralNetwork(train_X, train_Y, .5, [20, 10])

#    check the accuracy of our weights and biases 
neuralType = "Unregularized Neural Network"
checkAccuracy(w, b, train_X, train_Y, "Train", neuralType)
checkAccuracy(w, b, test_X, test_Y, "Test", neuralType)
```

### Decision Boundary
We finish by plotting the decision boundary for the unregularized Neural Network. 

```python
plot_decision_boundary(lambda x: predictPlotNeuralNetwork(w, b, x.T), train_X, C)
```

I believe this model to be overfitting as seen by some of the sharp cusps and curves in the decision boundary, especially towards the center of the plot. The overfitting can be explained by the model's lack of regularization.

# 3. Neural Network with L2 Parameter Regularization

We now train a 3-layer neural network with L2 regularization to classify the data - it has 2 hidden layers with 20 and 10 neurons respectively, uses ReLU in the hidden layers and sigmoid in the output layer. We plot the cost function, and calculate the training and test accuracy and error rates. 

```python
#   neural network with L2 regularization
def L2NeuralNetwork(X, y, alpha, lambd, layer_list):
    f, m = X.shape
    
    # initialize weights, biases and parameters 
    w = functools.reduce(lambda acc, neurons: (acc[0] + [np.random.randn(neurons, 
    acc[1]) * np.sqrt(2.0/acc[1])], neurons), layer_list, ([], f))[0]
    b = [ np.random.randn(neurons, 1) * 0.01 for neurons in layer_list ]
    z = [None] * len(w)
    a = [None] * len(w)
    costs = []

    for i in range(0, 2000):

        # forward propopgation
        z[0] = w[0] @ X + b[0]
        for i in range(len(w)-1):
            a[i] = ReLU(z[i])
            z[i+1] = w[i+1] @ a[i] + b[i+1]
        a[-1] = sigmoid(z[-1])

        # compute cost
        cost = - (y @ np.log(a[-1].T) + (1 - y) @ np.log(1 - a[-1].T))[0,0] / m
        
        # calculate L2 term
        L2 = 0
        for i in range(len(w)-1):
            L2 += np.sum(np.square(w[i]))
        L2 = L2 * (lambd/(2*m))
        
        # append L2 regularized cost to costs list for later plotting
        costs.append(cost + L2)


        dZ = a[-1] - y
        gradients = [] 
        # backward propogation 
        for i in range(len(z)-2, -1, -1):
            # calculate dW and dB with L2 term added to dW
            gradients.append([ ((dZ @ a[i].T) / m) + (lambd/m)*w[i+1], 
                np.sum(dZ, axis=1).reshape(b[i+1].shape) / m ])
        
            dZ = (w[i+1].T @ dZ) * dReLU(z[i])

        # calculate dW and dB for the first layer with L2 term added to dW
        gradients.append([ ((dZ @ X.T) / m) + (lambd/m)*w[0], 
            np.sum(dZ, axis=1).reshape(b[0].shape) / m ])

        # flip gradients so order is from first to last layer
        gradients = gradients[::-1]

        # update parameters 
        for j in range(len(gradients)):
            w[j] = w[j] - (alpha * gradients[j][0])
            b[j] = b[j] - (alpha * gradients[j][1])

    # plot cost function
    plt.clf()
    plt.plot(costs)
    plt.title("L2 Neural Network Cost Function (" + str(m) + " samples)")
    plt.xlabel("Number of Iterations")
    plt.ylabel("Cost")
    plt.show()
    return(w, b)

#    train our neural network and get back our weights and biases 
# learning rate = 0.5, L2 lambda = 0.1, hidden layers have 20 and 10 neurons 
w,b = L2NeuralNetwork(train_X, train_Y, 0.5, 0.1, [20, 10])

#    check the accuracy of our weights and biases 
neuralType = "L2 Neural Network"
checkAccuracy(w, b, train_X, train_Y, "Train", neuralType)
checkAccuracy(w, b, test_X, test_Y, "Test", neuralType)
```

L2 regularization has helped smoothen our decision boundary as compared with the decision boundary of the unregularized neural network which had more sharp curves and cusps.

# 4. Neural Network with Dropout 

We conclude by training a 3-layer neural network with Dropout regularization to classify the data - it has 2 hidden layers with 20 and 10 neurons respectively, uses ReLU in the hidden layers and sigmoid in the output layer. We plot the cost function, and calculate the training and test accuracy and error rates. 

```python
def dropout(a, drop, keep_probability):
    drop = np.random.rand(a.shape[0], a.shape[1])
    drop = drop < keep_probability
    a = np.multiply(a, drop)
    a = a / keep_probability
    return (a, drop)

def dropoutNeuralNetwork(X, y, alpha, keep_probability, layer_list):
    f, m = X.shape
    
    # initialize weights, biases and parameters 
    w = functools.reduce(lambda acc, neurons: (acc[0] + [np.random.randn(neurons, 
    acc[1]) * np.sqrt(2.0/acc[1])], neurons), layer_list, ([], f))[0]
    b = [ np.random.randn(neurons, 1) * 0.01 for neurons in layer_list ]
    z = [None] * len(w)
    a = [None] * len(w)
    drop = [None] * len(w)
    costs = []

    for i in range(0, 2000):

        # forward propopgation with dropout regularization
        z[0] = w[0] @ X + b[0]
        for i in range(len(w)-1):
            a[i] = ReLU(z[i])
            a[i], drop[i] = dropout(a[i], drop[i], keep_probability)
            z[i+1] = w[i+1] @ a[i] + b[i+1]
        a[-1] = sigmoid(z[-1])

        # compute cost
        cost = np.nan_to_num(- (y @ np.log(a[-1].T) + (1 - y) @ np.log(1 - a[-1].T))[0,0] / m)
        costs.append(cost)

        gradients = []
        dZ = a[-1] - y
        
        # backward propogation with dropout regularization 
        for i in range(len(z)-2, -1, -1):
            # calculate dW and dB
            gradients.append([ (dZ @ a[i].T) / m, 
                np.sum(dZ, axis=1).reshape(b[i+1].shape) / m ])
        
            # calculate dZ with dropout regularization 
            dZ = np.multiply((w[i+1].T @ dZ), drop[i]) * dReLU(z[i])
            dZ = dZ/keep_probability

        # calculate dW and dB for the first layer
        gradients.append([ (dZ @ X.T) / m, 
            np.sum(dZ, axis=1).reshape(b[0].shape) / m ])

        # flip gradients so order is from first to last layer
        gradients = gradients[::-1]

        # update parameters 
        for j in range(len(gradients)):
            w[j] = w[j] - (alpha * gradients[j][0])
            b[j] = b[j] - (alpha * gradients[j][1])

    # plot cost function
    plt.clf()
    plt.plot(costs)
    plt.title("Unregularized Neural Network Cost Function (" + str(m) + " samples)")
    plt.xlabel("Number of Iterations")
    plt.ylabel("Cost")
    plt.show()
    return(w, b)

#    train our neural network and get back our weights and biases 
# alpha = 0.5, keep_probability = 0.2, 2 hidden layers with 20 and 10 neurons 
w,b = dropoutNeuralNetwork(train_X, train_Y, 0.2, 0.5, [20, 10])

#    check the accuracy of our weights and biases 
neuralType = "Dropout Neural Network"
checkAccuracy(w, b, train_X, train_Y, "Train", neuralType)
checkAccuracy(w, b, test_X, test_Y, "Test", neuralType)
```

### Decision Boundary
We finish by plotting the decision boundary for the Dropout Neural Network. 

```python
plot_decision_boundary(lambda x: predictPlotNeuralNetwork(w, b, x.T), train_X, C)
```

I believe this model to be slightly overfitting. The training and test accuracy rates are almost the same, but the decision boundary has a couple of sharp cusps. Dropout helps with overfitting and we can see we are getting better/smoother results in this decision boundary as compared with the decision boundary of the unregularized neural network which had more sharp curves and cusps.
